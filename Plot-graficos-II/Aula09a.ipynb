## Gráfico de Linhas
Vamos criar um gráfico de linhas para analisar a expectativa de vida dos países da África.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("LifeExpectancy.csv")
```

```python
df.head()
```

```python
plt.rcParams["figure.figsize"] = [12,8]
sns.set_style("darkgrid")
df.plot(x="Countries", y = ["Males", "Females"])
plt.title("Comparando a expectativa de vida de homens e mulheres nos países da África")
plt.xlabel("Países da África")
plt.ylabel("Expectativa de vida")
plt.xticks(df.index,df['Countries'],rotation=90)
plt.show()
```

Vamos fazer um barplot com esses dados.

```python
plt.rcParams["figure.figsize"] = [12,8]
df.plot(x="Countries", kind="bar",stacked=False)
plt.title("Comparando a expectativa de vida de homens e mulheres nos países da África")
plt.xlabel("Países da África")
plt.ylabel("Expectativa de vida")
plt.show()
```
