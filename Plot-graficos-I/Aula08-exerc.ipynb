1) Utilizando o dataset 'tips.csv' crie um gráfico de área empilhado mostrando o faturamento total do restaurante por dia, exibindo o total gasto por homens e mulheres (duas pilhas).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("tips.csv")
```

```python
df.head()
```

```python
fat_dia = pd.crosstab(df['day'], df['sex'], values=df['total_bill'], aggfunc=np.sum)
fat_dia
```

```python
plt.rcParams["figure.figsize"] = [10,8]
sns.set_context("notebook", font_scale=1.5, rc={"font.size":14,"axes.titlesize":16,"axes.labelsize":12})
fat_dia.plot(kind="bar",stacked=False)
plt.title("Faturamento Homem x Mulher")
plt.ylabel("Faturamento")
plt.xlabel("Dia da semana")
plt.show()
```
